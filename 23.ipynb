```python
file = './23.txt'
lines = open(file, 'r').readlines()
lines = [l.strip('\n') for l in lines]
```

```python
computers = {}
t_computers = []
mappings = {}
n = 1
for line in lines:
    c1, c2 = line.split('-')
    if c1 not in computers:
        n1 = n
        computers[c1] = n1
        n += 1
        mappings[n1] = []
        if c1.startswith('t'):
            t_computers.append(n1)
    else:
        n1 = computers[c1]
    if c2 not in computers:
        n2 = n
        computers[c2] = n2
        n += 1
        mappings[n2] = []
        if c2.startswith('t'):
            t_computers.append(n2)
    else:
        n2 = computers[c2]
    if n1<n2:
        mappings[n1].append(n2)
    else:
        mappings[n2].append(n1)

print(len(computers))
print(len(t_computers))
print(len(mappings))
print(sum([len(m) for c, m in mappings.items()]))
computer_rev = {v: k for k, v in computers.items()}
```

```python
triplets = []
triplets_t = []
for n1 in range(1,len(mappings)+1):
    for n2 in mappings[n1]:
        thirds = set(mappings[n1]).intersection(mappings[n2])
        for n3 in thirds:
            triplets.append((n1,n2,n3))
            if n1 in t_computers or n2 in t_computers or n3 in t_computers:
                triplets_t.append((n1,n2,n3))

print(len(triplets))
print(len(triplets_t))
```
