```python
class Graph:
    def __init__(self, graph: dict = {}):
       self.graph = graph  # A dictionary for the adjacency list

    def min_distance(self, distances, sptSet):
        d=None
        minv=None
        for v in self.graph:
            if not sptSet[v] and (d is None or distances[v]<d):
                d = distances[v]
                minv = v
        return minv
    
    def dijkstra(self, src):
        assert src in self.graph
        distances = {node: float("inf") for node in self.graph}
        distances[src] = 0  # Set the source value to 0
        sptSet = {node: False for node in self.graph}
        for _ in tqdm.tqdm(range(len(self.graph))):

            u = self.min_distance(distances, sptSet)
            sptSet[u] = True
            for v in self.graph:
                if (v in self.graph[u]) and (sptSet[v] == False) and distances[v] > distances[u] + self.graph[u][v]:
                    distances[v] = distances[u] + self.graph[u][v]
        return distances

def build_graph(num_pixels):
    grid = np.zeros((71,71))
    for pix in info[:num_pixels]:
        grid[pix[1],pix[0]] = 1
    graph = {}
    for x in range(grid.shape[1]):
        for y in range(grid.shape[0]):
            if grid[y,x]>0:
                continue
            graph[(x,y)] = {}
            if y<grid.shape[0]-1 and grid[y+1,x]==0:
                graph[(x,y)][(x,y+1)]=1
            if x<grid.shape[1]-1 and grid[y,x+1]==0:
                graph[(x,y)][(x+1,y)]=1
            if y>0 and grid[y-1,x]==0:
                graph[(x,y)][(x,y-1)]=1
            if x>0 and grid[y,x-1]==0:
                graph[(x,y)][(x-1,y)]=1
    return graph

min_pixel = 1024
max_pixels = len(info)
found = False
found_pixels = None
tentatives = 0
search_range = (min_pixel, max_pixels)
while not found and tentatives<100:
    current_pixels = int((search_range[1]+search_range[0])/2)
    print(f'Building graph with {current_pixels} pixels')
    graph = build_graph(current_pixels)
    start_vertex = (0,0)
    end_vertexes = (70,70)
    graph_class = Graph(graph)
    res = graph_class.dijkstra(start_vertex)
    if search_range[1]-search_range[0]==1:
        found_pixels = search_range[0]
        found=True
    if res[end_vertexes]<100000:
        search_range = (current_pixels, search_range[1])
    else:
        search_range = (search_range[0], current_pixels)

    print(f'Result is {res[end_vertexes]}, so new range is {search_range}')
    tentatives+=1

current_pixels = found_pixels+1
print(f'Building graph with {current_pixels} pixels')
graph = build_graph(current_pixels)
start_vertex = (0,0)
end_vertexes = (70,70)
graph_class = Graph(graph)
res = graph_class.dijkstra(start_vertex)
print(f'Result is {res[end_vertexes]}, so new range is {search_range}')

print(f'FOUND RESULTS AT PIXEL {found_pixels}, with coordinates {info[found_pixels-1]}')
```
