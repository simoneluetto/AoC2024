```python
import numpy as np
import tqdm
import math
```

```python
file = './11.txt'
lines = open(file, 'r').readlines()
lines = [l.strip('\n') for l in lines]
print(len(lines))
```

```python
blinks = 25

arrangement = [int(n) for n in lines[0].split(' ')]
print(f'initial arrangment: {arrangement}')

for blink in tqdm.tqdm(range(blinks)):
    new_arrangment = []
    for n in arrangement:
        if n==0:
            new_arrangment.append(1)
        elif len(str(n))%2==0:
            s = str(n)
            c = int(len(s)/2)
            new_arrangment.append(int(s[:c]))
            new_arrangment.append(int(s[c:]))
        else:
            new_arrangment.append(n*2024)
    
    arrangement = new_arrangment.copy()
    print(f'blink {blink} -> \n{arrangement}')
    print(f'blink {blink} -> length = {len(arrangement)}')
            
print(f'FINAL after {blinks} blinks, number of stone is {len(arrangement)}')
```

```python
from functools import cache

@cache
def count_stone_blink(stone,blinks):
    if blinks==0:
        return 1
    else:
        if stone==0:
            return count_stone_blink(1,blinks-1)
        s = str(stone)
        ndig = len(s)
        if ndig%2==0:
            c = ndig//2
            return count_stone_blink(int(s[:c]),blinks-1) + count_stone_blink(int(s[c:]),blinks-1)
        # ndig = math.floor(math.log(stone+1,10))+1
        # if ndig%2==0:
        #     m = 10**(ndig//2)
        #     return count_stone_blink(n//m,blinks-1) + count_stone_blink(n%m,blinks-1)
        else:
            return count_stone_blink(stone*2024,blinks-1)

blinks = 75

arrangement = [int(n) for n in lines[0].split(' ')]
tot = 0
for stone in arrangement:
    tot += count_stone_blink(stone, blinks)

print(tot)
```

```python
from collections import defaultdict
found_res = defaultdict()

def comp_res_iterative(n, blinks):
    def comp_res(n):
        if n==0:
            return [1]
        elif len(str(n))%2==0:
            s = str(n)
            c = int(len(s)/2)
            return [int(s[:c]), int(s[c:])]
        else:
            return [n*2024]
        
    if n in found_res:
        if blinks in found_res[n]:
            return found_res[n][blinks]
        for found_blinks in range(blinks-1, 1, -1):
            if found_blinks in found_res[n]:
                return [el for r in found_res[n][found_blinks] for el in comp_res_iterative(r, blinks-found_blinks)]

    # print(f'called on {n}, {blinks}')
    if blinks==0:
        return [n]
    elif blinks>1:
        res = [el for r in comp_res(n) for el in comp_res_iterative(r, blinks-1)]
        if n not in found_res:
            found_res[n] = {}
        found_res[n][blinks] = res
        return res
    else:
        res = comp_res(n)
        if n not in found_res:
            found_res[n] = {}
        found_res[n][blinks] = res
        return res

blinks = 35

arrangement = [int(n) for n in lines[0].split(' ')]
nums_found = set()
results_found = {}
print(f'initial arrangment: {arrangement}')

final_res = comp_res_iterative(arrangement[0], blinks)
print(len(final_res))
# print(final_res)
# print(found_res)
```

```python
from collections import defaultdict

blinks = 75
blink_split = 15

starting_arrangements = [int(n) for n in lines[0].split(' ')]
nums_found = set()
results_found = {}
print(f'initial arrangment: {arrangement}')

def compute_res_iterative(elements, blinks, blink_split):
    def compute_res_blink_split(el, blink_split):
        curr_arrangement = [el]
        for _ in range(blink_split):
            new_arrangment = []
            for n in curr_arrangement:
                if n==0:
                    new_arrangment.append(1)
                elif len(str(n))%2==0:
                    s = str(n)
                    c = int(len(s)/2)
                    new_arrangment.append(int(s[:c]))
                    new_arrangment.append(int(s[c:]))
                else:
                    new_arrangment.append(n*2024)
                nums_found.add(n)
            curr_arrangement = new_arrangment.copy()
        return curr_arrangement

    num_splits = int(blinks/blink_split)
    curr_elements = elements.copy()
    for i in range(num_splits):
        new_elements = []
        for el in curr_elements:
            if el in results_found:
                res = results_found[el]
                # print(f'using precomputed res for el {el}, L {len(res)}')
            else:
                res = compute_res_blink_split(el, blink_split)
                results_found[el] = res
                # print(f'computed res for el {el}, L {len(res)}')
            new_elements.extend(results_found[el])
        curr_elements = new_elements.copy()
        print(f'done split {i}/{num_splits}')
    return new_elements

tot_res = compute_res_iterative(starting_arrangements, blinks, blink_split)
print(len(tot_res))
# print(results_found)
```
