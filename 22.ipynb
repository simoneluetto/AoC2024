```python
import numpy as np
import tqdm
```

```python
file = './22.txt'
lines = open(file, 'r').readlines()
lines = [l.strip('\n') for l in lines]
secrets = [int(l) for l in lines]
print(len(secrets))
```

```python
def next_secret(n):

    n = ((n*64)^n)%16777216
    n = (int(n/32)^n)%16777216
    n = ((n*2048)^n)%16777216
    return n

n = 123
for _ in range(10):
    n = next_secret(n)
    print(n)

ns = [1,10,100,2024]
for n in ns:
    initial_n = n
    for _ in range(2000):
        n = next_secret(n)
    print(f'2000th secret starting from {initial_n} -> {n}')

res = []
for n in tqdm.tqdm(secrets):
    initial_n = n
    for _ in range(2000):
        n = next_secret(n)
    res.append(n)
print(f'sum of 2000th secrets : {sum(res)}')
```

```python
prices = []
changes = []
all_secrets = []
for n in tqdm.tqdm(secrets):
    initial_n = n
    curr_price = n%10
    monk_prices = []
    monk_changes = []
    monk_secrets = []
    for _ in range(2000):
        n = next_secret(n)
        p = n%10
        monk_secrets.append(n)
        monk_prices.append(p)
        monk_changes.append(p-curr_price)
        curr_price = p
    res.append(n)
    prices.append(monk_prices)
    changes.append(monk_changes)
    all_secrets.append(monk_secrets)

print(f'starting of first monkey : {secrets[0]}')
print(f'secrets of first monkey : {all_secrets[0][:10]}')
print(f'prices of first monkey : {prices[0][:10]}')
print(f'changes of first monkey : {changes[0][:10]}')
```

```python
all_monkeys_changes_price = []

for i in range(len(secrets)):
    monkey_changes_price = {}
    monk_prices = prices[i]
    monk_changes = changes[i]
    for j in range(len(monk_changes)-1,2,-1):
        changes_sequence = (monk_changes[j-3],monk_changes[j-2],monk_changes[j-1],monk_changes[j])
        res_price = monk_prices[j]
        monkey_changes_price[changes_sequence] = res_price

    all_monkeys_changes_price.append(monkey_changes_price)

print(f'starting of first monkey : {secrets[0]}')
print(f'secrets of first monkey : {all_secrets[0][:10]}')
print(f'prices of first monkey : {prices[0][:10]}')
print(f'changes of first monkey : {changes[0][:10]}')
print(f'sequence pricing of the first monkey: {all_monkeys_changes_price[0]}')
```

```python
all_sequence_pricing = {}
all_sequence_times = {}
for monkey_changes_price in tqdm.tqdm(all_monkeys_changes_price):
    for seq, price in monkey_changes_price.items():
        if seq in all_sequence_pricing:
            all_sequence_pricing[seq] += price
            all_sequence_times[seq] += 1
        else:
            all_sequence_pricing[seq] = price
            all_sequence_times[seq] = 1

print(f'all sequence pricing : {all_sequence_pricing}')
all_sequence_pricing = [(v,k) for k,v in all_sequence_pricing.items()]
all_sequence_pricing = sorted(all_sequence_pricing, reverse=True)
print(f'all_sequence_pricing : {all_sequence_pricing}')
```
