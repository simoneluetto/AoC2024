```python
import numpy as np
import tqdm
import logging
import cv2
np.set_printoptions(edgeitems=100, linewidth=1000)
```

```python
file1 = './15_1.txt'
lines1 = open(file1, 'r').readlines()
lines1 = [l.strip('\n') for l in lines1]

file2 = './15_2.txt'
lines2 = open(file2, 'r').readlines()
lines2 = [l.strip('\n') for l in lines2]

print(len(lines1))
print(len(lines2))
```

```python
map_arr = []
moves = []
values_map_arr = {
    '.': 0,
    'O': 1,
    '@': 2,
    '#': 3,
}
for line in lines1:
    
    values = [values_map_arr[n] for n in line]
    map_arr.append(values)

for line in lines2:
    moves.extend(line)

map_arr = np.array(map_arr)
original_map_arr = map_arr.copy()

print(len(moves))
print(map_arr.shape)
print(np.unique(map_arr, return_counts=True))
print(map_arr[:20,:20])
```

```python
def robot_moves(map_arr, move):

    robot_y, robot_x = np.argwhere(map_arr==2)[0]
    
    if move=='^':
        new_pos_val = map_arr[robot_y-1,robot_x]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y-1,robot_x]=2
        elif new_pos_val==3:
            return map_arr
        else:
            found = False
            i = 1
            while not found:
                i+=1
                if map_arr[robot_y-i,robot_x]==3:
                    found='wall'
                elif map_arr[robot_y-i,robot_x]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y-1,robot_x]=2
                map_arr[robot_y-i,robot_x]=1
    elif move=='>':
        new_pos_val = map_arr[robot_y,robot_x+1]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y,robot_x+1]=2
        elif new_pos_val==3:
            return map_arr
        else:
            found = False
            i = 1
            while not found:
                i+=1
                if map_arr[robot_y,robot_x+i]==3:
                    found='wall'
                elif map_arr[robot_y,robot_x+i]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y,robot_x+1]=2
                map_arr[robot_y,robot_x+i]=1
    elif move=='<':
        new_pos_val = map_arr[robot_y,robot_x-1]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y,robot_x-1]=2
        elif new_pos_val==3:
            return map_arr
        else:
            found = False
            i = 1
            while not found and i<20:
                i+=1
                if map_arr[robot_y,robot_x-i]==3:
                    found='wall'
                elif map_arr[robot_y,robot_x-i]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y,robot_x-1]=2
                map_arr[robot_y,robot_x-i]=1
    elif move=='v':
        new_pos_val = map_arr[robot_y+1,robot_x]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y+1,robot_x]=2
        elif new_pos_val==3:
            return map_arr
        else:
            found = False
            i = 1
            while not found:
                i+=1
                if map_arr[robot_y+i,robot_x]==3:
                    found='wall'
                elif map_arr[robot_y+i,robot_x]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y+1,robot_x]=2
                map_arr[robot_y+i,robot_x]=1
    else:
        raise ValueError(move)
    # print(map_arr[robot_y-2:robot_y+3,robot_x-2:robot_x+3])
    
    return map_arr

map_arr = original_map_arr.copy()
for move in tqdm.tqdm(moves):
    map_arr = robot_moves(map_arr, move)
    # break

boxes = np.argwhere(map_arr==1)
res = sum([100*b[0]+b[1] for b in boxes])
res
```

```python
def robot_moves(map_arr, move, idx, debug=False):

    robot_y, robot_x = np.argwhere(map_arr==3)[0]
    # if debug:
    #     print(f'Start, move {idx}={move}, Rob:x={robot_x},y={robot_y}: \n{map_arr[robot_y-3:robot_y+4,robot_x-5:robot_x+6]}')
    cnt=0
    if move=='^':
        new_pos_val = map_arr[robot_y-1,robot_x]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y-1,robot_x]=3
        elif new_pos_val==4:
            return map_arr
        else:
            found = False
            i = 0
            to_check = (robot_x, robot_x+1) if new_pos_val==1 else (robot_x-1,robot_x)
            to_check_by_i = {0: to_check}

            if debug:
                print(f'MOVE : {idx}')
                print(f'before pushing up for i={i}, \n{map_arr[robot_y-8:robot_y+2,robot_x-4:robot_x+5]}')

            while not found:
                i+=1
                map_to_check = map_arr[robot_y-i,to_check[0]:to_check[1]+1]
                if debug:
                    print(f'trying to push up, checking row {map_to_check}')
                if map_to_check.max()==4:
                    found='wall'
                elif map_to_check.max()==0:
                    found='space'
                else:
                    if map_to_check[0]==2:
                        to_check = (to_check[0]-1,to_check[1])
                        map_to_check = map_arr[robot_y-i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'enlarging left map to check for boxes to {map_to_check}')
                        cnt += 1
                    if map_to_check[-1]==1:
                        to_check = (to_check[0],to_check[1]+1)
                        map_to_check = map_arr[robot_y-i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'enlarging right map to check for boxes to {map_to_check}')
                        cnt += 1
                    while map_to_check[0]==0:
                        to_check = (to_check[0]+1,to_check[1])
                        map_to_check = map_arr[robot_y-i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'reducing left map to check for zeros to {map_to_check}')
                        cnt += 1
                    while map_to_check[-1]==0:
                        to_check = (to_check[0],to_check[1]-1)
                        map_to_check = map_arr[robot_y-i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'reducing right map to check for zeros to {map_to_check}')
                        cnt += 1

                to_check_by_i[i] = to_check
            if found=='wall':
                if debug:
                    print(f'cannot push, wall found')
                return map_arr
            else:
                if debug:
                    print(f'pushing with widths : {to_check_by_i}')
                map_arr[robot_y,robot_x]=0
                for j in range(i,1,-1):
                    map_arr[robot_y-j,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1] = np.where(
                        map_arr[robot_y-j+1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1]>0,
                        map_arr[robot_y-j+1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1],
                        map_arr[robot_y-j,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1])
                    map_arr[robot_y-j+1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1] = 0
                # map_arr[robot_y-i:robot_y-1,to_check[0]:to_check[1]+1] = map_arr[robot_y-i+1:robot_y,to_check[0]:to_check[1]+1]
                map_arr[robot_y-1,to_check_by_i[0][0]:to_check_by_i[0][1]+1]=0
                map_arr[robot_y-1,robot_x]=3
                if debug:
                    print(f'pushed, \n{map_arr[robot_y-8:robot_y+2,robot_x-4:robot_x+5]}')
    elif move=='v':
        new_pos_val = map_arr[robot_y+1,robot_x]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y+1,robot_x]=3
        elif new_pos_val==4:
            return map_arr
        else:
            found = False
            i = 0
            to_check = (robot_x, robot_x+1) if new_pos_val==1 else (robot_x-1,robot_x)
            to_check_by_i = {0: to_check}
            if debug:
                print(f'MOVE : {idx}')
                print(f'before pushing down for i={i}, \n{map_arr[robot_y-1:robot_y+9,robot_x-4:robot_x+5]}')

            while not found:
                i+=1
                map_to_check = map_arr[robot_y+i,to_check[0]:to_check[1]+1]
                if debug:
                    print(f'trying to push down, checking row {map_to_check}')
                if map_to_check.max()==4:
                    found='wall'
                elif map_to_check.max()==0:
                    found='space'
                else:
                    if map_to_check[0]==2:
                        to_check = (to_check[0]-1,to_check[1])
                        map_to_check = map_arr[robot_y+i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'enlarging left map to check for boxes to {map_to_check}')
                        cnt += 1
                    if map_to_check[-1]==1:
                        to_check = (to_check[0],to_check[1]+1)
                        map_to_check = map_arr[robot_y+i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'enlarging right map to check for boxes to {map_to_check}')
                        cnt += 1
                    while map_to_check[0]==0:
                        to_check = (to_check[0]+1,to_check[1])
                        map_to_check = map_arr[robot_y+i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'reducing left map to check for zeros to {map_to_check}')
                        cnt += 1
                    while map_to_check[-1]==0:
                        to_check = (to_check[0],to_check[1]-1)
                        map_to_check = map_arr[robot_y+i,to_check[0]:to_check[1]+1]
                        if debug:
                            print(f'reducing right map to check for zeros to {map_to_check}')
                        cnt += 1
                to_check_by_i[i] = to_check
            if found=='wall':
                if debug:
                    print(f'cannot push, wall found')
                return map_arr
            else:
                if debug:
                    print(f'pushing with widths : {to_check_by_i}')
                map_arr[robot_y,robot_x]=0
                for j in range(i,1,-1):
                    map_arr[robot_y+j,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1] = np.where(
                        map_arr[robot_y+j-1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1]>0,
                        map_arr[robot_y+j-1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1],
                        map_arr[robot_y+j,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1])
                    map_arr[robot_y+j-1,to_check_by_i[j-1][0]:to_check_by_i[j-1][1]+1] = 0
                # map_arr[robot_y+2:robot_y+i+1,to_check[0]:to_check[1]+1] = map_arr[robot_y+1:robot_y+i,to_check[0]:to_check[1]+1]
                map_arr[robot_y+1,to_check_by_i[0][0]:to_check_by_i[0][1]+1]=0
                map_arr[robot_y+1,robot_x]=3
                if debug:
                    print(f'pushed, \n{map_arr[robot_y-1:robot_y+9,robot_x-4:robot_x+5]}')

    elif move=='>':
        new_pos_val = map_arr[robot_y,robot_x+1]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y,robot_x+1]=3
        elif new_pos_val==4:
            return map_arr
        else:
            found = False
            i = 1
            while not found:
                i+=1
                if map_arr[robot_y,robot_x+i]==4:
                    found='wall'
                elif map_arr[robot_y,robot_x+i]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                # if debug:
                #     print(f'before pushing right for i={i}, \n{map_arr[robot_y-2:robot_y+3,robot_x-2:robot_x+5]}')
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y,robot_x+2:robot_x+i+1] = map_arr[robot_y,robot_x+1:robot_x+i]
                map_arr[robot_y,robot_x+1]=3
                # if debug:
                #     print(f'pushed, \n{map_arr[robot_y-2:robot_y+3,robot_x-2:robot_x+5]}')
    elif move=='<':
        new_pos_val = map_arr[robot_y,robot_x-1]
        if new_pos_val==0:
            map_arr[robot_y,robot_x]=0
            map_arr[robot_y,robot_x-1]=3
        elif new_pos_val==4:
            return map_arr
        else:
            found = False
            i = 1
            while not found:
                i+=1
                if map_arr[robot_y,robot_x-i]==4:
                    found='wall'
                elif map_arr[robot_y,robot_x-i]==0:
                    found='space'
            if found=='wall':
                return map_arr
            else:
                # if debug:
                #     print(f'before pushing left for i={i}, \n{map_arr[robot_y-2:robot_y+3,robot_x-4:robot_x+3]}')
                map_arr[robot_y,robot_x]=0
                map_arr[robot_y,robot_x-i:robot_x-1] = map_arr[robot_y,robot_x-i+1:robot_x]
                map_arr[robot_y,robot_x-1]=3
                # if debug:
                #     print(f'pushed, \n{map_arr[robot_y-2:robot_y+3,robot_x-4:robot_x+3]}')
    else:
        raise ValueError(move)

    # if debug:
    #     print(f'Moved, move {idx}={move}: \n{map_arr[robot_y-3:robot_y+4,robot_x-5:robot_x+6]}')
    # if cnt >= 8:
    #     print(f'on move {idx} enlarged/reduced more than 8 times -> {cnt}')

    return map_arr

new_map = original_new_map.copy()

moves_to_debug = [
# 1592,
# 1595,
# 1903,
# 2123,
# 2281,
# 11774,
# 11915,
# 11922,
# 12256,
# 12259,
# 12262,
# 12269,
# 16531,
# 16534,
# 19870,
# 19874,
# 19875,
# 19876
]

for idx, move in tqdm.tqdm(enumerate(moves)):
    try:
        if idx in moves_to_debug:
            new_map = robot_moves(new_map.copy(), move, idx, debug=True)
        else:
            new_map = robot_moves(new_map, move, idx, debug=False)
        for p in np.argwhere(new_map==1):
            if not new_map[p[0],p[1]+1]==2:
                print(f'after step {idx}, move {move} a box is Wrong \n{new_map[p[0]-2:p[0]+3,p[1]-2:p[1]+3]}')
                assert False
        for p in np.argwhere(new_map==2):
            if not new_map[p[0],p[1]-1]==1:
                print(f'after step {idx}, move {move} a box is Wrong \n{new_map[p[0]-2:p[0]+3,p[1]-2:p[1]+3]}')
                assert False
        if (new_map[0,:]==4).sum()<new_map.shape[1]:
            print(f'after step {idx}, move {move} the border up is wrong!')
            assert False
        if (new_map[-1,:]==4).sum()<new_map.shape[1]:
            print(f'after step {idx}, move {move} the border down is wrong!')
            assert False
        if (new_map[:,0]==4).sum()<new_map.shape[0]:
            print(f'after step {idx}, move {move} the border left is wrong!')
            assert False
        if (new_map[:,-1]==4).sum()<new_map.shape[0]:
            print(f'after step {idx}, move {move} the border right is wrong!')
            assert False
    except:
        assert False
    # break

boxes = np.argwhere(new_map==1)
res = sum([100*b[0]+b[1] for b in boxes])
res
```
