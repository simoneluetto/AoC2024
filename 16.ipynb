```python
import numpy as np
import tqdm
```

```python
file = './16.txt'
lines = open(file, 'r').readlines()
lines = [l.strip('\n') for l in lines]
```

```python
maze = []
mapping = {'.':0,'#':1,'S':2,'E':3}
for line in lines:
    values = [mapping[n] for n in line]
    maze.append(values)

maze = np.array(maze)
print(maze.shape)
print(np.unique(maze,return_counts=True))
```

```python
rotate_clockwise_dict={
    'E':'S',
    'S':'W',
    'W':'N',
    'N':'E',
}
rotate_counterclockwise_dict={
    'E':'N',
    'N':'W',
    'W':'S',
    'S':'E',
}
```

```python
start_pos = np.argwhere(maze==2)[0]
end_pos = np.argwhere(maze==3)[0]

def next_pos(v):
    if v[2]=='E':
        return (v[0],v[1]+1,v[2])
    elif v[2]=='W':
        return (v[0],v[1]-1,v[2])
    elif v[2]=='S':
        return (v[0]+1,v[1],v[2])
    elif v[2]=='N':
        return (v[0]-1,v[1],v[2])
    else:
        raise ValueError()
def rotate_clockwise(f):
    return rotate_clockwise_dict[f]
def rotate_counterclockwise(f):
    return rotate_counterclockwise_dict[f]

vertices = [(start_pos[0],start_pos[1],'E')]
current_vertex = [(start_pos[0],start_pos[1],'E')]
edges = []
completed=False
while not completed:
    new_vertices = []
    for v in current_vertex:
        v1 = next_pos(v)
        if maze[v1[0],v1[1]] in [0,3]:
            if v1 not in vertices:
                vertices.append(v1)
                new_vertices.append(v1)
            if (v,v1,1) not in edges:
                edges.append((v,v1,1))
        v2 = (v[0],v[1],rotate_clockwise(v[2]))
        if v2 not in vertices:
            vertices.append(v2)
            new_vertices.append(v2)
        if (v,v2,1000) not in edges:
            edges.append((v,v2,1000))
        v3 = (v[0],v[1],rotate_counterclockwise(v[2]))
        if v3 not in vertices:
            vertices.append(v3)
            new_vertices.append(v3)
        if (v,v3,1000) not in edges:
            edges.append((v,v3,1000))              
    completed = len(new_vertices)==0
    current_vertex = new_vertices
```

```python
print(len(vertices))
print(len(edges))
```

```python
graph = {}
for vertex in vertices:
    graph[vertex] = {}
    for edge in edges:
        if edge[0]==vertex:
            graph[vertex][edge[1]]=edge[2]

print(graph)
```

```python
class Graph:
    def __init__(self, graph: dict = {}):
       
       self.graph = graph  # A dictionary for the adjacency list


    def min_distance(self, distances, sptSet):
        d=None
        minv=None
        for v in self.graph:
            if not sptSet[v] and (d is None or distances[v]<d):
                d = distances[v]
                minv = v
        return minv
    
    def dijkstra(self, src):
 

        assert src in self.graph
        distances = {node: float("inf") for node in self.graph}
        paths = {node: [node] for node in self.graph}
        distances[src] = 0  # Set the source value to 0
        sptSet = {node: False for node in self.graph}
 
        for _ in tqdm.tqdm(range(len(self.graph))):

            u = self.min_distance(distances, sptSet)
            sptSet[u] = True
 
            # Update dist value of the adjacent vertices
            # of the picked vertex only if the current
            # distance is greater than new distance and
            # the vertex in not in the shortest path tree
            for v in self.graph:
                if (v in self.graph[u]) and (sptSet[v] == False):
                    if distances[v] > distances[u] + self.graph[u][v]:
                        distances[v] = distances[u] + self.graph[u][v]
                        paths[v] = paths[u] + [v]
                    elif distances[v] == distances[u] + self.graph[u][v]:
                        paths[v] = paths[u] + paths[v]

        return distances, paths

start_vertex = (start_pos[0],start_pos[1],'E')
end_vertex = (end_pos[0],end_pos[1],'N')
graph_class = Graph(graph)

res, paths = graph_class.dijkstra(start_vertex)

print(res[end_vertex], paths[end_vertex])
```
